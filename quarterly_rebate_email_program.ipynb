```python
import pandas as pd
import numpy as np
```

```python
# Create functions to remove ("+", "-", ":", ",") from columns

def dashes_to_zero(array):
    array = array.replace('-', 0)
    return array       

def string_to_float(array):
    array = (array.str.replace("+", "").
                   str.replace("%", "").
                   str.replace(",", "").
                   str.replace(":", "").
                   str.strip().
                   replace(r'^\s*$', 0, regex=True).
                   astype(float)
            )
    return array
```

# Import H2_2020 Data and Clean

```python
h2_cols = ['code', 'customer', 'Q3_units', 'Q3_dollars', 'Q4_units', 'Q4_dollars']
h2 = pd.read_csv("2020_H2_Data.csv", names = h2_cols)
h2 = h2.drop([0,1,2])
```

```python
h2 = h2.reset_index(drop=True)
```

```python
h2_features_to_clean = ['Q3_units', 'Q3_dollars', 'Q4_units', 'Q4_dollars']

h2[h2_features_to_clean] = h2[h2_features_to_clean].apply(dashes_to_zero)
h2[h2_features_to_clean] = h2[h2_features_to_clean].apply(string_to_float)
```

```python
h2 = h2.fillna(0)
```

```python
h2['Q4_2020_total_units'] = (h2['Q3_units'] + h2['Q4_units']).astype(int)
h2['Q4_2020_total_dollars'] = h2['Q3_dollars'] + h2['Q4_dollars']
h2.head()
```

# Alternate method - rebate data using pivot

```python
# new column names
columns = ["code", "customer", "city", "year", "quarter", "units", "dollars"]

# read data with new column names
pivot_df = pd.read_csv("210525_Q2_Rebate_Pivot.csv", header = 1, names = columns)
# drop totals top row and reset index
# pivot_df = pivot_df.drop(0)
# pivot_df = pivot_df.reset_index(drop=True)
pivot_df
```

```python
pivot_df_units = pivot_df[["code", "year", "units"]]
pivot_df_units = pivot_df_units.pivot(index = "code", columns = "year", values = "units")
```

```python
flattened_units = pd.DataFrame(pivot_df_units.to_records())
flattened_units = flattened_units.rename(columns = {'2019': '2019_units', '2020': '2020_units', '2021': '2021_units'})
flattened_units = flattened_units.fillna(0)
flattened_units
```

```python
pivot_df_dollars = pivot_df[["code", "year", "dollars"]]
pivot_df_dollars = pivot_df_dollars.pivot(index = "code", columns = "year", values = "dollars")
```

```python
flattened_dollars = pd.DataFrame(pivot_df_dollars.to_records())
flattened_dollars = flattened_dollars.rename(columns = {'2019': '2019_dollars', '2020': '2020_dollars', '2021': '2021_dollars'})
flattened_dollars = flattened_dollars.fillna(0)
flattened_dollars
```

```python
merged_2 = pd.merge(left = flattened_units, right = flattened_dollars, how = "left", left_on = "code", right_on = "code")
merged_2
```
