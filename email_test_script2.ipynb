```python
import smtplib
import sys
sys.path.append('/Users/matt/Desktop/DS/Oakley/Rebate_Email_Project/Config')
import config
import time
import pandas as pd

from email.message import EmailMessage
from datetime import date

today = date.today()
print(today)
```

```python
def send_email(subject, msg, customer_email):
    try:
        email = EmailMessage()
        email['Subject'] = subject
        email['From'] = config.EMAIL_ADDRESS
        email['To'] = customer_email
        email.set_content(msg, subtype='html')
        
        server = smtplib.SMTP('smtp.gmail.com:587')
        server.ehlo()
        server.starttls()
        server.login(config.EMAIL_ADDRESS, config.PASSWORD)
        server.send_message(email)
        server.quit()
        print(f'Success. Email Sent to {customer_email}')
        time.sleep(1)
    except:
        print(f'Email Failed to Send to {customer_email}')
```
